```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
from spacy import displacy
from spacy import tokenizer
from IPython.display import HTML, display
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bbc_data = pd.read_csv('bbc_news.csv')
```

```python
bbc_data.head()
```

```python
bbc_data.info()
```

```python
titles = pd.DataFrame(bbc_data["title"])
titles.head()
```

```python
titles["title_lowercase"] = titles["title"].str.lower()
titles.head()
```

```python
en_stopwords = stopwords.words('english')
titles["no_stopwords"] = titles["title_lowercase"].apply(lambda x: " ".join([word for word in x.split() if word not in en_stopwords]))
```

```python
titles["no_stopwords_no_punct"] = titles["no_stopwords"].apply(lambda x: re.sub(r"[^\w\s]","",x))
```

```python
titles.head()
```

```python
nltk.download('punkt_tab')
```

```python
titles["tokens_raw"] = titles["title"].apply(lambda x: word_tokenize(x))
```

```python
titles["tokens_raw_clean"] = titles["no_stopwords_no_punct"].apply(lambda x: word_tokenize(x))
```

```python
titles.head()
```

```python
lemmatizer = WordNetLemmatizer()
titles["tokens_clean_lemmatized"] = titles["tokens_raw_clean"].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
titles.head()
```

```python
tokens_raw_list = sum(titles["tokens_raw"], [])
tokens_clean_list = sum(titles["tokens_clean_lemmatized"], [])
```

```python
print(tokens_clean_list)
print(tokens_raw_list)
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
spacy_doc = nlp(" ".join(tokens_raw_list))
```

```python
pos_df = pd.DataFrame(columns=["token", "pos_tag"])
for word in spacy_doc:
    pos_df = pd.concat([pos_df, pd.DataFrame.from_records([{"token": word.text, "pos_tag": word.pos_}])], ignore_index=True)
```

```python
print(pos_df.head(10))
```

```python
pos_df_counts = pos_df.groupby(['token', 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

```python
pos_df_counts.head(10)
```

```python
nouns = pos_df_counts[pos_df_counts['pos_tag'] == 'NOUN'][0:10]
```

```python
nouns
```

```python
verbs = pos_df_counts[pos_df_counts["pos_tag"] == "VERB"][0:10]
```

```python
verbs
```

```python
ner_df = pd.DataFrame(columns=["token", "ner_tag"])

for word in spacy_doc.ents:
    if not pd.isna(word.label_):
        ner_df = pd.concat([ner_df, pd.DataFrame.from_records([{"token": word.text, "ner_tag": word.label_}])], ignore_index=True)
```

```python
ner_df.head(10)
```

```python
ner_df_counts = ner_df.groupby(["token", "ner_tag"]).size().reset_index(name="counts").sort_values(by="counts", ascending=False)
```
